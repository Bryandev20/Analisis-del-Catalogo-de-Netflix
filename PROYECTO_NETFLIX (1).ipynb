🔑 Reglas prácticas

* <5% nulos 👉 imputar con moda o valores cercanos.

* 5–30% nulos 👉 depende del valor: si es importante, imputar; si no, puedes poner "Unknown".

* Mayor de 30% nulos 👉 evaluar si realmente aporta valor. Si no, eliminar columna o imputar con "Unknown".
* Mayor de 75%  nulos 👉 Se debe eliminar la columna

```python
df_netflix_limpio=df_netflix.copy()

df_netflix_limpio['director']=df_netflix_limpio['director'].fillna('Unknown')
df_netflix_limpio['cast']=df_netflix_limpio['cast'].fillna('Unknown')
df_netflix_limpio['country']=df_netflix_limpio['country'].fillna(df_netflix_limpio['country'].mode()[0]) #Moda
```

```python
# date_added, rating y duration tienen poco porcentaje de valores nulos
df_netflix_limpio.dropna(subset=['date_added','rating','duration'], inplace=True)
```

```python
#Duplicados
df_netflix_limpio.drop_duplicates(inplace=True)
```

```python
df_netflix_limpio.info()
```

```python
df_netflix_limpio.head()
```

```python
#convertir el formato texto a año
df_netflix_limpio['date_added'] = pd.to_datetime(df_netflix_limpio['date_added'], errors='coerce')
```

```python
df_netflix_limpio['date_added'] = df_netflix_limpio['date_added'].dt.year.astype('Int64')
df_netflix_limpio['date_added'] = df_netflix_limpio['date_added'].astype(str).replace('<NA>', 'unknown')
```

```python
#Eliminar columnas
columnas=['cast','description']
df_netflix_limpio.drop(columns=columnas,inplace=True)
```

```python
#Seleccionar la columna listed solo el primer genero

df_netflix_limpio['listed_in']=df_netflix_limpio['listed_in'].str.split(', ').str[0]
df_netflix_limpio=df_netflix_limpio.explode('listed_in').reset_index(drop=True)
df_netflix_limpio.head()
```
